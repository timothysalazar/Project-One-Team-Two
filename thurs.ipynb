```python
data = response.json()
```

```python
df = pd.DataFrame(data["result"])
```

```python
df.loc[4, "habitat"] = "Forest - Subtropical/Tropical Moisttt Montane"
df
```

```python
def get_category(row):
    return row.replace(" ","").split("-")[0]
    
df["category"] = df.habitat.map(get_category)
df
```

```python
df.category.value_counts().plot.bar()
```
