```python
import json 
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
path = "/Users/grahamjaeger/um/Data/united-states-fish-species-threatened.csv"
df = pd.read_csv(path, encoding='latin-1')
print(df)
```

```python
print(df.columns)
```

```python
threat_counts = df['Threat Category'].value_counts()
print(threat_counts)
```

```python
occurrence_counts = df['Occurrence'].value_counts()
print(occurrence_counts)
```

```python

contingency_table = pd.crosstab(index=df['Occurrence'], columns=df['Threat Category'])
print(contingency_table)

```

```python
contingency_table.plot(kind='bar', stacked=True)

# Adding labels and title
plt.xlabel('Occurrence')
plt.ylabel('Count')
plt.title('Occurrence vs Threat Category')

# Displaying the bar graph
plt.show()
```

```python
from scipy.stats import chi2_contingency

chi2, p_value, dof, expected = chi2_contingency(contingency_table)
print("Chi-square statistic:", chi2)
print("p-value:", p_value)
```
